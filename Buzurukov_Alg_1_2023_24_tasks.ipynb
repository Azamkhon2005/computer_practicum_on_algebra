###  Задание 3.
Упростите выражение
$$
\frac{(u+v)^2-4uv}{u-v}+2v-1
$$

```python
u,v = symbols("u, v")
simplify(((u + v)**2 - 4*u*v ) / (u-v) + 2*v-1 )
```

###  Задание 4.

```python
a1,a2,a3 = symbols("a1, a2, a3")
expand((a1 +a2+a3)**3-(a1-a2)**2)
```

###  Задание 5.
Разложите на множители
$2u^4+3u^3v-2u^2v^2-2u^2v-3uv^2+2v^3$

```python
u, v = symbols("u, v")
exp = 2*u**4 + 3*u**3*v-2*u**2*v**2-2*u**2*v-3*u*v**2+2*v**3
factor(exp)
```

###  Задание 6.
Сгруппируйте выражение из Задания 5 по степеням $u$.

```python
collect(exp,u)
```

### Задание 7
Создайте список $Powers$ из трех кортежей, состоящих из символов $x$, $y$ и $z$ в степени, равной номеру кортежа в списке.

```python
powers = [symbols("x,y,z"), symbols("x,y,z"), symbols("x,y,z")]
for i in range(len(powers)):
  prep = list(elem**i for elem in powers[i])
  powers[i] = tuple(prep)
```

## Матрицы
### Задание 9
Составить матрицу  $A_{3\times 4}$ из чисел вида $3k+1$, $k=1,2,...$:
$$
A=
\left(
\begin{matrix}
4& 7& 10& 13\\
16& 19& 22& 25\\
28& 31& 34& 37
\end{matrix}
\right)
$$

```python
k = 1
arr = []
for i in range(3):
  arr.append([0]*4)
  for j in range(4):
    arr[i][j] = 3*k + 1
    k += 1
m = Matrix(arr)
m
```

### Задание 10
В матрице А задания 9 удалить вторую строку и вставить третий столбец между первым и вторым (сам третий столбец не удалять!), должно получиться
$$
A=
\left(
\begin{matrix}
4& 10& 7& 10& 13\\
28& 34& 31& 34& 37
\end{matrix}
\right)
$$

```python
m.row_del(1)
m = m.col_insert(1,m.col(2))
m
```
