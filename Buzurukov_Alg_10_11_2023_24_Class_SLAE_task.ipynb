**Дочерний класс SLAEhomogeneous**

**переменные** класса:

*  set_b этот метод  нужно переопределить, чтобы он ничего не возвращал и не делал, а только выводил сообщение 'b = 0 in homogeneous SLAE, use get_b instead'

*  x метод проверяет, что матрица квадратная и невырожденная, возвращает в этом случае кортеж с True и нулевым вектором соответствующей размерности, не использует np.linalg.solve,
иначе False и пустой np.array



**Дочерний класс SLAEsquare**

**переменные** класса:

singular: True если вырожденная матрица, иначе False

square: True если квадратная матрица, иначе False

a_inv обратная матрица, если существует, иначе None

**атрибуты**

is_square возвращает square, при необходимости предварительно вычисляя и заполняя эту переменную

is_singular - возвращает singular, при необходимости предварительно вычисляя и заполняя эту переменную

get_inv - - возвращает обратную матрицу, если она существует

**метод**


x проверяет невырожденность матрицы методами этого же класса, для невырожденной матрицы возвращает True и решение или False и  пустой np.array

```python
@dataclass
class SLAE:
    name: str
    a: np.array
    b: np.array
    var_num:int = field(default=0, repr=False)
    eq_num:int = field(default=0, repr=False)

    @property
    def get_a(self) -> np.array:
      return self.a

    @property
    def get_b(self) -> np.array:
      return self.b

    @property
    def get_var_num(self) -> int:
      if self.var_num == 0:
        self.var_num = len(self.a[0])
      return self.var_num

    @property
    def get_eq_num(self) -> int:
      if self.eq_num == 0:
        self.eq_num = len(self.a)
      return self.eq_num

    @property
    def get_dim(self) -> tuple:
      return (self.eq_num, self.get_var_num)

    def set_b(self, b:np.array):
      if len(b) == self.a.shape[0]:
        self.b = b

    def set_b_zero(self):
      self.b = np.zeros(self.a.shape[0])

    def x(self, solver = np.linalg.solve) -> tuple:
      a_matr = self.get_a
      n, m = self.get_a.shape
      res = np.array([])
      if n == m and np.linalg.det(a_matr) != 0:
          res =  solver(a_matr, self.get_b)
      else:
        return (False, res)
      if numpy.linalg.matrix_rank(a_matr) == np.append(a_matr, self.get_b, axis=1) == n:
        return (True, res)
      else:
        return (False, np.array([]))
```

```python
@dataclass
class SLAEhomogeneous(SLAE):
  b: np.array = field(default=None, repr=False)

  @property
  def get_b(self)-> np.array:
    return np.zeros(len(self.a))

  def set_b(self):
    print("b = 0 in homogeneous SLAE, use get_b instead.")

  def x(self) -> tuple:
    n, m = self.get_a.shape
    if n == m and np.linalg.det(self.get_a) !=0 :
      return (True, np.zeros(n))
    else:
      return (False, np.array([]))
```

```python
@dataclass
class SLAEsquare(SLAE):
  singular:bool = field(default=None, repr=False)
  square:bool = field(default=None, repr=False)
  a_inv:np.array = field(default=None, repr=False)


  @property
  def is_square(self) -> bool:
    if self.square is None:
      self.set_square()
    return self.square

  @property
  def is_singular(self) -> bool:
    if self.singular is None:
      self.set_singular()
    return self.singular

  @property
  def get_inv(self) -> np.array:
    if self.a_inv is not None:
      return self.a_inv

  def set_singular(self):
    if np.linalg.det(self.get_a)==0:
      self.singular = True
    else:
      self.singular = False
  def set_square(self):
    if self.get_a.shape[0] == self.get_a.shape[1]:
      self.square = True
    else:
      self.square = False
  def set_inv(self):
    try:
      self.a_inv = np.linalg.inv(self.a)
    except:
      self.a_inv = None


  def x(self, solver = np.linalg.solve) -> tuple:
    if self.singular is False and self.square and (self.a_inv is not None):
      res =  solver(self.a, self.b)
      return (True, res)
    return (False, np.array([]))
```
