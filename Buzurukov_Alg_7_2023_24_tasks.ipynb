### Задание 1.
Даны точки: $A(3, -2, 3)$,  $B(-2, 3, 1)$, $C(1, -8, 6)$, $D(1, 2, 4)$, $F(-7, 8, -1)$.

Составить словарь точек (ключи - точки, значения - имена точек).

Вывести на экран значения записей словаря (т.е. имена точек).

Проверить, различны ли прямые $AB$ и $AF$, $AB$ и $AC$, плоскости $ABC$ и $ACF$, $ABC$ и $ACD$.

Вывести на экран текст вида "$AB$ и $AF$ одинаковы" (или "не одинаковы").

```python
s = {Point(3,-2,3):"A",
     Point(-2, 3, 1): "B",
     Point(1, -8, 6): "C",
     Point(1, 2, 4): "D",
     Point(-7, 8, -1): "F"}

# Вывести на экран значения записей словаря (т.е. имена точек).
print(s.values())


#Проверить, различны ли прямые  AB  и  AF ,  AB  и  AC , плоскости  ABC  и  ACF ,  ABC  и  ACD .
# Неудобно работать со словарем, в котором ключи-точки а значения-имена точек. Поэтому лучше реализовать словарь в котором ключи-имена точек а значение-сами точки.
Points = {"A": Point(3,-2,3),
     "B":Point(-2, 3, 1),
     "C":Point(1, -8, 6),
     "D":Point(1, 2, 4),
     "F":Point(-7, 8, -1)}
AB = Line(Points["A"], Points["B"])
AC = Line(Points["A"], Points["C"])
AF = Line(Points["A"], Points["F"])
if AB.is_similar(AC):
  print("AB и AC одинаковы")
else:
  print("AB и AC не одинаковы")

if AB.is_similar(AF):
  print("AB и AF одинаковы")
else:
  print("AB и AF не одинаковы")
print("--"*20)
ABC = Plane(Points["A"], Points["B"], Points["C"])
ACF = Plane(Points["A"], Points["C"], Points["F"])
ACD = Plane(Points["A"], Points["C"], Points["D"])

if ABC.is_coplanar(ACF):
  print("ABC и ACF одинаковы")
else:
  print("ABC и ACF не одинаковы")

if ABC.is_coplanar(ACD):
  print("ABC и ACD одинаковы")
else:
  print("ABC и ACD не одинаковы")
```
